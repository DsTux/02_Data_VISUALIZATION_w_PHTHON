```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
age = [25, 27, 29, 31, 33, 35, 37, 39, 41, 43, 45]
salary = [38496, 42000, 46752, 49320, 53200,
         56000, 62316, 64928, 67317, 68748, 73752]
```

```python
plt.plot(age, salary)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary by Age')
```

```python
salary_2 = [45372, 48876, 53850, 57287, 63016,
            65998, 70003, 70000, 71496, 75370, 83640]
```

```python
plt.plot(age, salary, label='salary')
plt.plot(age, salary_2, label='salary_2')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary by Age')
plt.legend();
```

### Subplot cizme Fonksionel yontem ile

```python
# Functional yontemde subplot cizme
plt.subplot(2,1,1)
plt.plot(age,salary, "r")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.subplot(1,2,1)
plt.plot(age,salary, "r")
plt.subplot(2,1,2)
plt.plot(age, salary_2, "b")
plt.tight_layout();
```

```python
plt.subplot(2,2,1)
plt.plot(age,salary, "r")
plt.subplot(2,2,2)
plt.plot(age, salary, "b")
plt.subplot(2,2,3)
plt.plot(age,salary, "g")
plt.subplot(2,2,4)
plt.plot(age,salary, "y")
plt.tight_layout();
```

```python
row, col, places = 2,2,4
colors = ["r","b","g","y"]
for place in range(places):
    plt.subplot(row,col,place+1)
    plt.plot(age,salary,colors[place])
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0,0.0,1,1])
ax2 = fig.add_axes([0.9,0.3,0.4,0.4])

ax1.plot(age,salary,'r');
ax1.set_xlabel('Age')
ax1.set_ylabel('Salary')
ax1.set_title('Salary by Age')

ax2.plot(age,salary,'y');
ax2.set_xlabel('Age')
ax2.set_ylabel('Salary_2')
ax2.set_title('Salary_2 by Age')
```
